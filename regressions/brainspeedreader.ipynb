```python
def AR1(n,x0,c,phi,sigma=1):   
    X = [x0]
    for i in range(n):
        X.append(c + phi * X[-1] + np.random.normal(scale=sigma))
        #theta*(mu-X[-1]) + sigma * )
    return X

def randWordFreq(baseline,std,n):
    return AR1(n,baseline,baseline/2,0.5,sigma=std)
```

```python
figure(1,(15,5))
x = randWordFreq(250,50,500)
plot(x,'+-')
np.mean(x),np.std(x),S.skew(x),S.kurtosis(x)

```
